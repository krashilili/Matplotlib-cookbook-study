```python
# dependencies
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plotting curves
x = range(100)
y1 = [value**2 for value in x]
y2 = [value for value in x]
```

```python
plt.plot(x,y1)
plt.plot(x,y2)
plt.show()
```

```python
# plot points
data = np.random.rand(1024,2)
plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
# plot bars
data = [[5., 25., 50., 20.,],
        [4., 23., 51., 17.,],
        [6., 22., 52., 19.]]
```

```python
x = np.arange(4)
plt.bar(x+0.00, data[0], color='b', width=.25)
plt.bar(x+0.25, data[1], color='g', width=.25)
plt.bar(x+0.50, data[2], color='r', width=.25)
plt.show()
```

```python
# plot stacked bars
y1 = np.array([5., 30., 45., 22.])
y2 = np.array([5., 25., 50., 20.])
x_axis = range(4)
plt.bar(x_axis, y1, color='b')
plt.bar(x_axis, y2, color='r', bottom=y1)
plt.show()
```

```python
# plot back-to-back bars
plt.barh(x_axis, y1, color='r')
plt.barh(x_axis, 0-y2, color='b')
plt.show()
```
